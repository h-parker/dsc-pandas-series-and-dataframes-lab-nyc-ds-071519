## Group the Data by Day of Week and Plot the Sum of The Numeric Columns

```python
df['DATE'] = pd.to_datetime(df.DATE, format='%m/%d/%Y').dt.day_name()
groupby_date = df.groupby('DATE').sum()
groupby_date
```

```python
groupby_date.index
```

## Group the Data by Weekend/Weekday and Plot the Sum of the Numeric Columns

```python
mapping = {'Saturday': 'Weekend', 'Sunday': 'Weekend', 'Monday': 'Weekday',
           'Tuesday': 'Weekday', 'Wednesday': 'Weekday', 'Thursday': 'Weekday',
           'Friday': 'Weekday'}
groupby_weekday_or_end = groupby_date.groupby(mapping).sum()
groupby_weekday_or_end
```
